```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
URL="https://kitabay.com/collections/fiction-romance"
```

```python
page = requests.get(URL)
```

```python
page.status_code
```

```python
htmlcode = page.text
soup = BeautifulSoup(htmlcode)
```

```python
htmlcode
```

```python
#title
title = soup.find('div',attrs={'class':'product__title__wrapper'})
print(title.text)
```

```python
#price
price = soup.find('span',attrs={'class':'product__price--sale'})
price.text
```

```python
#rating
Rating = soup.find('div',attrs={'class':'ratingWrapper'})
Rating.text
```

```python
#description
desc = soup.find('div',attrs={'class':'accordion-content__inner'})
desc.text
```

```python
#other details of book
other_details = soup.find('div',attrs={'class':'accordion__wrapper'})
other_details.text
```

```python
#total details of one product
total_details = soup.find('div',attrs={'class':'product__inner is-sticky'})
total_details.text
```

```python
x = soup.find_all('div',attrs={'class':'collection__content'})
for i in x:
  print(i.text)
```

```python
title = []
authors = []
for i in range(1,10):
  URL="https://kitabay.com/blogs/second-hand-books"
  page=requests.get(URL)
  htmlcode=page.text
  soup=BeautifulSoup(htmlcode)
  for x in soup.find_all('div',attrs={'class':'blog section-padding'}):
    items=x.find('div',attrs={'class':'article__item-content'})
    if items is None:
      title.append(np.NaN)
    else:
      title.append(items.text)
    author=x.find('div',attrs={'class':'article__item-author'})
    if author is None:
      authors.append(np.NaN)
    else:
      authors.append(author.text)
```

```python
print(len(title))
print(len(authors))
```

```python
df=pd.DataFrame({'items':title,'author':authors})
df.head()
```
